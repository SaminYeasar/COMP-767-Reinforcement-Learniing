## (d)Application, where regret optimization would be more useful than best arm identification.


###################################################

When reward function of arms become a function of time regret optimization would be more useful. Total regret in cumalative over all time steps. When we do too much of exploration at the begining, even if we end up at optimal arm regret will be high. So when we have limited number of trials and we want our explore and exploit to be in a balanced way regret optimization would be best suited.
